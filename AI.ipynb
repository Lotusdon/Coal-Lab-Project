```python
maze = [
 ['S', 0, 0, 1, 0],
 [1, 0, 1, 0, 0],
 [0, 0, 0, 0, 'G'],
 [1, 1, 0, 1, 1]
]

rows = len(maze)
cols = len(maze[0])

def find_pos(value):
  for i in range(rows):
    for j in range(cols):
      if maze[i][j] == value:
        return (i,j)

start = find_pos('S')
goal = find_pos('G')

def valid(x,y):
  return x>=0 and x<rows and y>=0 and y<cols and maze[x][y] != 1

def dls(node, depth, limit, visited, path):
  if node == goal:
    return path

  if depth == limit:
    return None

  x,y = node
  moves = [(x+1,y),(x-1,y),(x,y+1),(x,y-1)]

  for nx,ny in moves:
    if valid(nx,ny) and (nx,ny) not in visited:
      visited.append((nx,ny))
      result = dls((nx,ny), depth+1, limit, visited, path+[(nx,ny)])
      if result:
        return result

  return None

def depth_limited_search(limit):
  visited = [start]
  return dls(start,0,limit,visited,[start])

def iterative_deepening(max_depth):
  for depth in range(max_depth+1):
    result = depth_limited_search(depth)
    if result:
      return result
  return None

print("Depth Limited Search (limit=6):")
print(depth_limited_search(6))

print("Iterative Deepening Search:")
print(iterative_deepening(10))
```
